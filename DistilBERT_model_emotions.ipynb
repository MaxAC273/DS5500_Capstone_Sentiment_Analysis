```python
from transformers import pipeline

# Initialize the pipeline for sentiment analysis

#nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
nlp = pipeline("sentiment-analysis", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
```

```python
# Making predictions
def get_highest_scoring_label(text):
    # Analyze the text and get all scores - adjust based on actual structure
    all_scores = nlp(text)
    # Flatten the list of lists to a single list of dictionaries
    # Assuming there's a misunderstanding and the structure is not as deeply nested as thought
    scores = [score for sublist in all_scores for score in sublist]
    # Find the highest scoring emotion across all returned dictionaries
    highest_score = max(scores, key=lambda score: score['score'])
    # Return the label of the highest scoring emotion
    return highest_score['label']

# Apply the function to each row in the DataFrame
df['sentiment'] = df['cleaned_text'].apply(get_highest_scoring_label)

#df['sentiment'] = df['cleaned_text'].apply(lambda x: nlp(x)[0]['label'])
```
